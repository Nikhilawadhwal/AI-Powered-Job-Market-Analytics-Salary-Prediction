# Import ML Libaries

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
```

## *Feature Engineering*

### *Unnecessary Columns Remove*

```python
df.drop(['Job_ID'], axis=1, inplace=True)
```

### *Check Data Types*

```python
df.dtypes
```

```python
print(df['Experience'].unique())
```

```python
df['Experience'] = df['Experience'].str.strip()
```

```python
df['Experience'].isna().sum()
```

```python
df['Experience'].value_counts(dropna=False)
```

#### *Experience Convert*

```python
Experience_map = {
    'Fresher':0,
    '0-2':1,
    '1-3':2,
    '2-5':3,
    '3-6':4,
    '5+':5
}

df['Experience'] = df['Experience'].map(Experience_map)
```

```python
df.info()
```

```python
df['Experience']
```

## *Feature Selection*   

```python
X = df.drop('Salary', axis=1)
y = df['Salary']

print("Features Shape :", X.shape)
print("Target Shape :", y.shape)
```

## *Identify Categorical Features*

```python
X.select_dtypes(include='object').columns
```

## *Encode Categorical Variables*

```python
le = LabelEncoder()

categorical_columns = [
'Job_Title',
'Company',
'Location',
'Skills',
'Education',
'Job_Type',
'Department',
'Industry',
'Status'
]

for col in categorical_columns:
    X[col] = le.fit_transform(X[col])

X.head()
```

## *Check Missing Values*

```python
X.isnull().sum()
```

## *Handle Missing Values*

```python
X.fillna(X.median(numeric_only=True), inplace=True)
```

## *Split the Dataset into Training and Testing Sets*

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print("Training Features :", X_train.shape)
print("Testing Features :", X_test.shape)
print("Training Target :", y_train.shape)
print("Testing Target :", y_test.shape)
```

## *Standardize the Features*

```python
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)

X_test = scaler.transform(X_test)
```

## *Train the Linear Regression Model*

```python
lr_model = LinearRegression()

lr_model.fit(X_train, y_train)
```

## *Predict Salary*

```python
y_pred = lr_model.predict(X_test)

y_pred[:10]
```

## *Evaluate the Performance of the Linear Regression Model*

```python
mae = mean_absolute_error(y_test, y_pred)

mse = mean_squared_error(y_test, y_pred)

rmse = np.sqrt(mse)

r2 = r2_score(y_test, y_pred)

print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
print("R² Score:", r2)
```

## *Compare Actual and Predicted Salary*

```python
comparison = pd.DataFrame({
    "Actual Salary": y_test.values,
    "Predicted Salary": y_pred
})

comparison.head(10)
```

## *Visualize the Actual vs Predicted Salary*

```python
plt.figure(figsize=(10,6))

plt.scatter(y_test, y_pred, alpha=0.7)

plt.xlabel("Actual Salary")
plt.ylabel("Predicted Salary")
plt.title("Actual Salary vs Predicted Salary")

plt.show()
```

## *Train a Decision Tree Regressor*

```python
dt_model = DecisionTreeRegressor(random_state=42)

dt_model.fit(X_train, y_train)
```

```python
dt_pred = dt_model.predict(X_test)
dt_pred
```

## *Evaluate the Decision Tree Model*

```python
dt_mae = mean_absolute_error(y_test, dt_pred)
dt_mse = mean_squared_error(y_test, dt_pred)
dt_rmse = np.sqrt(dt_mse)
dt_r2 = r2_score(y_test, dt_pred)

print("MAE :", dt_mae)
print("MSE :", dt_mse)
print("RMSE:", dt_rmse)
print("R2 Score:", dt_r2)
```

## *Train a Random Forest Regressor*

```python
rf_model = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)

rf_model.fit(X_train, y_train)
```

```python
rf_pred = rf_model.predict(X_test)
print(rf_pred)
```

## *Evaluate the Random Forest Model*

```python
rf_mae = mean_absolute_error(y_test,rf_pred)
rf_mse =  mean_squared_error(y_test,rf_pred)
rf_rmse = np.sqrt(rf_mse)
rf_r2 = r2_score(y_test,rf_pred)

print("Mean Absolute Error (MAE):", rf_mae)
print("Mean Squared Error (MSE):", rf_mse)
print("Root Mean Squared Error (RMSE):", rf_rmse)
print("R² Score:", rf_r2)
```

## *Compare All Machine Learning Models*

```python
comparison = pd.DataFrame({
    "Model": [
        "Linear Regression",
        "Decision Tree",
        "Random Forest"
    ],
    "MAE": [
        mae,
        dt_mae,
        rf_mae
    ],
    "RMSE": [
        rmse,
        dt_rmse,
        rf_rmse
    ],
    "R2 Score": [
        r2,
        dt_r2,
        rf_r2
    ]
})

comparison.sort_values(
    by="R2 Score",
    ascending=False
)
```

## *Identify the Most Important Features*

```python
importance = pd.DataFrame({
    'Feature': X.columns,
    'Importance': rf_model.feature_importances_
})

importance = importance.sort_values(
    by='Importance',
    ascending=False
)

importance
```

# Encode Categorical Columns

```python
le = LabelEncoder()

for col in df.select_dtypes(include="object").columns:
    df[col] = le.fit_transform(df[col])
```

# Feature Scaling

```python
scaler = StandardScaler()

scaled_data = scaler.fit_transform(df)
```

# Elbow Method

```python
wcss = []

for i in range(1,11):
    km = KMeans(n_clusters=i,
                init='k-means++',
                random_state=42,
                n_init=10)

    km.fit(scaled_data)
    wcss.append(km.inertia_)
```

# Plot Elbow Curve

```python
plt.figure(figsize=(8,5))

plt.plot(range(1,11),wcss,marker='o')

plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.title("Elbow Method")

plt.show()
```

# Train KMeans

```python
kmeans = KMeans(
    n_clusters=4,
    init='k-means++',
    random_state=42,
    n_init=10
)

cluster = kmeans.fit_predict(scaled_data)
```

# Add Cluster Column

```python
df["Cluster"] = cluster

df.head()
```

# Cluster Count

```python
centers = pd.DataFrame(
    scaler.inverse_transform(kmeans.cluster_centers_),
    columns=df.columns[:-1]
)

centers
```

# Silhouette Score

```python
score = silhouette_score(scaled_data,cluster)

print("Silhouette Score :",score)
```

# PCA Visualization

```python
pca = PCA(n_components=2)

pca_data = pca.fit_transform(scaled_data)
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    pca_data[:,0],
    pca_data[:,1],
    c=cluster,
    cmap="viridis"
)

plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.title("Job Market Clusters")

plt.colorbar()

plt.show()
```

# UI Inter face in GRADIO

```python
import gradio as gr
import joblib
import pandas as pd

# Load Model
model = joblib.load("salary_model.pkl")

# Prediction Function
def predict_salary(experience, skills, education, company, location, job_title):

    input_data = pd.DataFrame({
        "Experience": [experience],
        "Skills": [skills],
        "Education": [education],
        "Company": [company],
        "Location": [location],
        "Job_Title": [job_title]
    })

    prediction = model.predict(input_data)

    return f"Predicted Salary: ₹ {prediction[0]:,.2f}"

# UI
demo = gr.Interface(
    fn=predict_salary,
    inputs=[
        gr.Number(label="Experience"),
        gr.Number(label="Number of Skills"),
        gr.Number(label="Education (Encoded)"),
        gr.Number(label="Company (Encoded)"),
        gr.Number(label="Location (Encoded)"),
        gr.Number(label="Job Title (Encoded)")
    ],
    outputs=gr.Textbox(label="Salary Prediction"),
    title="AI-Powered Job Market Analytics & Salary Prediction",
    description="Enter job details to predict salary."
)

demo.launch()
```
